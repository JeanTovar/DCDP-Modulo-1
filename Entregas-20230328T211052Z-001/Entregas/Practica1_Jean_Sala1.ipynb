### Función suma

$$
    suma(a) = \sum_{i=1}^{n}a_i
$$

````python
def suma(x):
    suma = 0
    i=0
    for i in x:
      #print("Entrada:", i)
      suma=suma+i
      #print("Suma acumulada", suma)
    return suma
````

````python
suma([1,2,3,4,5])
````

````python
assert suma([1, 2, 3, 4, 5]) == 15
````

### Función promedio.

$$
	\bar{x} = \dfrac{\sum_{i=1}^{n}x_i}{n}
$$

````python
def promedio(x):
    promedio = 0
    n=len(x)
    for i in x:
      promedio = suma(x)/n
      #print ("La suma es", suma)
      #print("el numero de elementos n es", n)
    return promedio
````

````python
promedio([1,2,3,4,5])
````

````python
assert promedio([1, 2, 3, 4, 5]) == 3, "Error en promedio"
````

### Producto punto (inner)

Completa la función inner que recibe como parametros dos listas a, b de la misma longitúd y retorna el producto punto de ambos.

El producto punto está definido como:

$$a = [a_1 a_2 ... a_n] \: y \: b = [b_1 b_2 ... b_n]$$

$$
a \cdot b = a_1*b_1 + a_2*b_2 + \dots + a_n*b_n
$$

````python
def inner(a, b):
  i=0
  ab=0
  while i< len(a):
    abi=a[i]*b[i]
    ab=ab+abi
    i=i+1
  return ab
    
````

````python
assert inner([1, 2, 3], [1, 2 ,3]) == 14, "Error en inner"
print("Correcto")
````

### Matriz identidad
$$
	I_{ij} = \begin{cases}
		1 & i=j\\
		0 & i \neq j\\
	\end{cases}
$$

````python
 def eye(n): 
  I = []
  for i in range(n):
      I.append([])
      for j in range(n):
          if i==j:
            I[i].append(1)
          else:  
            I[i].append(0)
  return (I)
````

````python
eye(3)
````

````python
assert eye(2) == [[1, 0], [0, 1]], "Error en eye"
````

### Matriz transpuesta

$$C = A^T \Leftrightarrow c_{ij} = a{ji}$$

Por ejemplo, si:
	$$
	A = \begin{bmatrix}
	a_{11} & a_{12}  \\
	a_{21} & a_{22}  \\
	a_{31} & a_{32}  \\
	\end{bmatrix}
	$$
	entonces
	$$
	A^T = \begin{bmatrix}
	a_{11} & a_{21} & a_{31} \\
	a_{12} & a_{22} & a_{32} \\
	\end{bmatrix}
	$$

````python

def traspose(A):
  A_t = [[0 for i in range(len(A))] for j in range(len(A[0]))]
  for i in range(len(A[0])):
    for j in range(len(A)):
      A_t[i][j] = A[j][i]
  return A_t
````

````python
A=[[1,2],[3,4],[5,6]]
traspose(A)


````

````python
assert traspose([[1, 2], [3, 4], [5, 6]]) == [[1, 3, 5], [2, 4, 6]], "Error en traspose"
````
